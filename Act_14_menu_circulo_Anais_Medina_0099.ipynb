## Redacción del Problema
Un estudiante de geometría necesita una herramienta interactiva para automatizar el cálculo de propiedades de figuras circulares. El objetivo es desarrollar un script en Python que, mediante un menú elegante, permita calcular:

El área de un círculo (espacio bidimensional).

El perímetro o circunferencia (longitud del contorno).

El volumen de una esfera (espacio tridimensional generado al rotar el círculo).

Fórmulas Matemáticas
Para realizar estos cálculos, utilizaremos las siguientes expresiones matemáticas:

Área del círculo: A=π⋅r

```python
import math

# --- Funciones de Cálculo ---

def areacirculo():
    print("\n--- 📐 Calculando Área ---")
    r = float(input("Introduce el radio del círculo: "))
    area = math.pi * (r ** 2)
    print(f"✨ El área es: {area:.2f} unidades²")

def perimetrocirculo():
    print("\n--- 📏 Calculando Perímetro ---")
    r = float(input("Introduce el radio del círculo: "))
    perimetro = 2 * math.pi * r
    print(f"✨ El perímetro es: {perimetro:.2f} unidades")

def volumencirculo():
    print("\n--- 🎈 Calculando Volumen de la Esfera ---")
    r = float(input("Introduce el radio para la esfera: "))
    volumen = (4/3) * math.pi * (r ** 3)
    print(f"✨ El volumen es: {volumen:.2f} unidades³")

# --- Función Principal / Menú ---

def menu():
    continuar = True

    while continuar:
        print("\n" + "═"*30)
        print("    🌟 GEOMETRÍA INTERACTIVA 🌟")
        print("═"*30)
        print("1️⃣. Calcular área del círculo")
        print("2️⃣. Calcular perímetro del círculo")
        print("3️⃣. Calcular volumen (Esfera)")
        print("4️⃣. Salir ❌")
        print("═"*30)

        opcion = input("Selecciona una opción (1-4): ")

        if opcion == '1':
            areacirculo()
        elif opcion == '2':
            perimetrocirculo()
        elif opcion == '3':
            volumencirculo()
        elif opcion == '4':
            print("\n¡Gracias por usar el programa! 👋✨")
            print("Programa realizado por anais medina NL = 36")
            continuar = False
        else:
            print("\n⚠️ Opción no válida, intenta de nuevo.")


# --- Ejecución del Programa ---
menu()
```
